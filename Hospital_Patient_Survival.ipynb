# Basic EDA

```python
# explore dataset

import numpy as np
import pandas as pd

train = pd.read_csv('dataset/training_data.csv')
train.head()
```

```python
train.isnull().sum()
```

```python
train.dtypes
```

```python
# explore distinct values for each column

for col in train:
  print(col, train[col].unique(), '\n')
```

```python
# extract relevant features

features = train[[
             'Diagnosed_Condition', 'Treated_with_drugs', 'Patient_Age',
             'Patient_Body_Mass_Index', 'Patient_Smoker', 'Patient_Rural_Urban',
             'A', 'B', 'C', 'D', 'E', 'F', 'Z',
            'Number_of_prev_cond', 'Survived_1_year'     
]].copy()
```

```python
def normalize_text(row):
  if pd.isnull(row):
    return row
  row = row.lower()
  row = row.strip()
  if row == 'yess':
    row = 'yes'
  return row
```

```python
# normalize ambiguous columns

features['Patient_Smoker'] = features['Patient_Smoker'].apply(normalize_text)
features['Treated_with_drugs'] = features['Treated_with_drugs'].apply(normalize_text)
```

```python
# check normalization result

for col in features:
  print(col, features[col].unique(), '\n')
```

```python
numerical_cols = [
                  'Diagnosed_Condition', 'Patient_Age', 'Patient_Body_Mass_Index',
                  'A', 'B', 'C', 'D', 'E', 'F', 'Z', 'Number_of_prev_cond'
]
```

```python
categorical_cols = [
                'Treated_with_drugs', 'Patient_Smoker', 'Patient_Rural_Urban'  
]
```

```python
# use KNNImputer and scaling for numerical missing values
# use mode value along with OneHotEncoder for categorical values 

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler

numerical_transformer = Pipeline(steps=[
    ('imputer', KNNImputer(weights='distance')),
    ('scaler', MinMaxScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encode', OneHotEncoder())
])

# define preprocessing steps

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
# split data into train and test sets
all_features = numerical_cols + categorical_cols
from sklearn.model_selection import train_test_split
x = features[all_features]
y =  features['Survived_1_year']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=0)
```

```python
from sklearn.metrics import f1_score
from catboost import CatBoostClassifier

# Build pipeline and fit data

my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', CatBoostClassifier(random_state=0))
                             ])

 
my_pipeline.fit(x_train, y_train)
```

```python
# get accuracy scores

my_pipeline.score(x_train, y_train), my_pipeline.score(x_test, y_test)
```

```python
# get F1 Score (contest metric)

preds = my_pipeline.predict(x_test)

score = f1_score(y_test, preds)
print(score)
```
